```python
def defence_control(team):
    country_data = test_data[
        (test_data["team_bat"] == team) &
        (test_data["date"] > pd.Timestamp("2018-09-20"))
    ]

    bazball_data = country_data[country_data["date"] > pd.Timestamp("2022-05-28")]
    pre_bazzball = country_data[country_data["date"] < pd.Timestamp("2022-05-28")]

    defence_pre = pre_bazzball[
        (pre_bazzball['shot'].isin(["DEFENDED", "LEFT_ALONE"])) &
        (pre_bazzball['bat'] != "Jo Root")
    ]
    control_defence_pre = defence_pre['control'].mean()
    pre_count = (
    pre_bazzball[
        pre_bazzball["shot"].notna() &
        pre_bazzball["control"].notna()
    ]["p_match"]
    .nunique()
)

    defence_post = bazball_data[
        (bazball_data['shot'].isin(["DEFENDED", "LEFT_ALONE"])) &
        (bazball_data['bat'] != "Jo Root")
    ]
    post_count = (
    bazball_data[
        bazball_data["shot"].notna() &
        bazball_data["control"].notna()
    ]["p_match"]
    .nunique()
)
    control_defence_post = defence_post['control'].mean()
    improvement = control_defence_post - control_defence_pre


    return {
        "Team": team,
        "Defence Control Pre-Bazball": round(control_defence_pre*100,2),
        "Pre-Bazball Matches": pre_count,
        "Defence Control Post-Bazball": round(control_defence_post*100,2),
        "Post-Bazball Matches": post_count,
        "Î” Control": round(improvement*100,2)
    }
```

```python
teams = [
    "England", "India", "Australia", "South Africa",
    "Sri Lanka", "West Indies", "New Zealand",
    "Pakistan", "Afghanistan", "Zimbabwe", "Ireland"
]
results = pd.DataFrame([defence_control(team) for team in teams])
display(results)
```

```python
import pandas as pd

# 1. Filter conditions
kohli_out_no_control = test_data[
    (test_data["bat"] == "Virat Kohli") &
    (test_data["out"] == 1) &
    #(test_data["control"] == 0) &
    (test_data["shot"].notna())
]

# 2. Shot counts
shot_counts = kohli_out_no_control["shot"].value_counts()

# 3. Shot percentages
shot_percent = kohli_out_no_control["shot"].value_counts(normalize=True) * 100

# 4. Combine into a table
kohli_shot_distribution = (
    pd.DataFrame({
        "Count": shot_counts,
        "Percentage": shot_percent.round(2)
    })
    .reset_index()
    .rename(columns={"index": "Shot"})
)

# 5. Display
display(kohli_shot_distribution)
```

```python
import pandas as pd

# Filter Kohli data with valid shot & control
kohli = test_data[
    (test_data["bat"] == "Virat Kohli") &
    (test_data["shot"].notna()) &
    (test_data["control"].notna())
]

# Group by shot type
summary = kohli.groupby("shot").agg(
    Total_Balls=("shot", "count"),
    Outs=("out", "sum"),
    Bat_Runs=("batruns", "sum"),
    No_Control=("control", lambda x: (x == 0).sum())
).reset_index()

# Percentages and rates
summary["Out_%"] = (summary["Outs"] / summary["Total_Balls"] * 100).round(2)
summary["No_Control_%"] = (summary["No_Control"] / summary["Total_Balls"] * 100).round(2)

# Strike Rate = runs per 100 balls
summary["Strike_Rate"] = (
    summary["Bat_Runs"] / summary["Total_Balls"] * 100
).round(2)

# Sort by usage (change if needed)
summary = summary.sort_values("Total_Balls", ascending=False)

display(summary)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

chrome_options = Options()
chrome_options.add_argument("--headless=new")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--window-size=1920,1080")

driver = webdriver.Chrome(options=chrome_options)

driver.get("https://www.iplt20.com/match/2025/1872")

wait = WebDriverWait(driver, 20)

# Click Scorecard
scorecard_btn = wait.until(
    EC.element_to_be_clickable((By.XPATH, "//a[@data-id='scoreCard']"))
)
driver.execute_script("arguments[0].click();", scorecard_btn)

# Wait for scorecard content
wait.until(
    EC.presence_of_element_located(
        (By.XPATH, "//div[contains(@class,'scoreCardContentWrap')]")
    )
)

html = driver.page_source
soup = BeautifulSoup(html, "lxml")

print(soup.prettify())

driver.quit()
```

```python
import asyncio
from playwright.async_api import async_playwright
from bs4 import BeautifulSoup

async def get_scorecard_html():
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()

        await page.goto(
            "https://www.iplt20.com/match/2025/1872",
            timeout=60000
        )

        await page.click("a[data-id='scoreCard']")
        await page.wait_for_selector("div.scoreCardContentWrap")

        html = await page.content()
        soup = BeautifulSoup(html, "lxml")
        scorecard_html = soup.select_one("div.scoreCardContentWrap")

        await browser.close()
        return scorecard_html

scorecard_html = await get_scorecard_html()
```
